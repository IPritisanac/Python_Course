## Variables and Data Types
Let's explore different data types and how to use them.

```python
x = 10  # Integer
y = 3.14  # Float
name = "Python"  # String
is_learning = True  # Boolean
print(f"x: {x}, y: {y}, name: {name}, is_learning: {is_learning}")
```

## Simple I/O and Control Flow
We can take input from users and make decisions using conditionals.

```python
number = int(input("Enter a number: "))
if number % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

## Working with Lists
Lists are one of the most commonly used data structures in Python.

```python
data = [5, 10, 15, 20]
data.append(25)
print(f"Updated list: {data}")

# Calculate the average of a list of numbers
def calculate_average(numbers):
    return sum(numbers) / len(numbers)

print(f"Average: {calculate_average(data)}")
```

## File Input/Output
Learn how to read from and write to files.

```python
# Writing to a file
with open("example.txt", "w") as file:
    file.write("Hello, Python!\n")
    file.write("This is an example file.")

# Reading from a file
with open("example.txt", "r") as file:
    content = file.read()
    print(content)
```

## Data Visualization
Use Matplotlib to create basic visualizations.

```python
import matplotlib.pyplot as plt
data = [10, 20, 15, 25, 30]
plt.plot(data, marker='o')
plt.title("Simple Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
