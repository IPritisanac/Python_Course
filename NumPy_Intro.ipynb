# NumPy Basics
This notebook introduces NumPy concepts like array creation, indexing, manipulation, and more.

```python
import numpy as np

# 1. Creating arrays
array_1d = np.array([10, 20, 30, 40])
array_2d = np.array([[1, 2, 3], [4, 5, 6]])
zeros = np.zeros((3, 3))
ones = np.ones((3, 3))
arange_example = np.arange(10)
print("1D Array:", array_1d)
print("2D Array:", array_2d)
print("Zeros:", zeros)
print("Ones:", ones)
print("Range:", arange_example)
```

```python
# 2. Indexing and slicing
print("Indexing and Slicing:")
print(array_1d[1])  # Access second element
print(array_2d[0, 2])  # Access element at row 0, column 2
print(array_2d[:, 1])  # Slice column 1
```

```python
# 3. Reshaping and transposing
reshaped = np.arange(12).reshape(3, 4)
print("Reshaped Array:")
print(reshaped)
print("Transposed Array:")
print(reshaped.T)
```

```python
# 4. Matrix multiplication
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])
matrix_result = np.dot(matrix1, matrix2)
print("Matrix Multiplication:")
print(matrix_result)
```

```python
# 5. Boolean logic
array = np.array([10, 20, 30, 40, 50])
print("Boolean Indexing:")
print(array[array > 25])  # Elements greater than 25

print("Logical Operations:")
print((array > 15) & (array < 35))  # AND operation
```
